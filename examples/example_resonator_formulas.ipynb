# Import

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('../garfield/')
from Resonator_formula import *
```

# Impedance functions

## Single transverse resonator, Q scan

```python
Rs = 1e6
resonant_frequency = 1e9

freq = np.linspace(0, 2.5e9, 1000)
plt.figure()
for Q in np.arange(0.2, 0.8, 0.1):
    plt.plot(freq, Resonator_transverse_imp(freq, Rs, Q, resonant_frequency),
            label=f'{Q=:.2f}')
plt.grid()
plt.legend()
plt.xlabel("Frequency [Hz]")
plt.ylabel("Transverse Impedance [Ohm/m]")
```

## Single transverse resonator, wake length scan

```python
Rs = 1e6
Q = 500
resonant_frequency = 1e9

freq = np.linspace(5e8, 1.5e9, 1000)
plt.figure()
for w_l in np.geomspace(1e1, 1e4, 4):
    plt.plot(freq, Resonator_transverse_imp(freq, Rs, Q, resonant_frequency, wake_length=w_l),
            label=f'{w_l=:.0f}', alpha=0.8)
plt.xlim(resonant_frequency - 1e8, resonant_frequency + 1e8)
plt.grid()
plt.legend()
plt.xlabel("Frequency [Hz]")
plt.ylabel("Transverse Impedance [Ohm/m]")
```

## Multiple longitudinal resonators

```python
dict_resonators_parameters = {
    1: [1e6, 1, 1.5e9],
    2: [2e6, 5, 5e8]
}

freq = np.linspace(0, 2.5e9, 1000)
plt.figure()
plt.plot(freq, n_Resonator_longitudinal_imp(freq, dict_resonators_parameters))
plt.grid()
plt.xlabel("Frequency [Hz]")
plt.ylabel("Longitudinal Impedance [Ohm]")
```

# Wake functions

## Single transverse resonator, Q scan

```python
Rs = 1e6
resonant_frequency = 1e9

times = np.linspace(1e-12, 1e-9, 1000)

plt.figure()
for Q in np.arange(0.2, 0.8, 0.1):
    plt.plot(times, Resonator_transverse_wake(times, Rs, Q, resonant_frequency),
            label=f'{Q=:.2f}')
plt.grid()
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Transverse Wake Function [V/C/m]")
```
